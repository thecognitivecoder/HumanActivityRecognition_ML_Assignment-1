```python
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This script combines the data from the UCI HAR Dataset into a more usable format.
# The data is combined into a single csv file for each subject and activity. 
# The data is then stored in the Combined folder.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
import pandas as pd
import numpy as np
import os

# Give the path of the test and train folder of UCI HAR Dataset
train_path = "C:\\Users\\thipp\OneDrive\Desktop\\human+activity+recognition+using+smartphones\\UCI HAR Dataset\\UCI HAR Dataset\\train"
test_path = "C:\\Users\\thipp\\OneDrive\\Desktop\\human+activity+recognition+using+smartphones\\UCI HAR Dataset\\UCI HAR Dataset\\test"

# Dictionary of activities. Provided by the dataset.
ACTIVITIES = {
    1: 'WALKING'            ,
    2: 'WALKING_UPSTAIRS'   ,
    3: 'WALKING_DOWNSTAIRS' ,
    4: 'SITTING'            ,
    5: 'STANDING'           ,
    6: 'LAYING'             ,
}

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                        # Combining Traing Data
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Load all the accelerometer data
total_acc_x = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_x_train.txt"),delim_whitespace=True,header=None)
total_acc_y = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_y_train.txt"),delim_whitespace=True,header=None)
total_acc_z = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_z_train.txt"),delim_whitespace=True,header=None)


# Read the subject IDs
subject_train = pd.read_csv(os.path.join(train_path,"subject_train.txt"),delim_whitespace=True,header=None)

# Read the labels
y = pd.read_csv(os.path.join(train_path,"y_train.txt"),delim_whitespace=True,header=None)


# Toggle through all the subjects.
for subject in np.unique(subject_train.values):

    sub_idxs = np.where( subject_train.iloc[:,0] == subject )[0]
    labels = y.loc[sub_idxs]

    # Toggle through all the labels.
    for label in np.unique(labels.values):

        # make the folder directory if it does not exist
        if not os.path.exists(os.path.join("Combined","Train",ACTIVITIES[label])):
            os.makedirs(os.path.join("Combined","Train",ACTIVITIES[label]))

        label_idxs = labels[labels.iloc[:,0] == label].index

        accx = []
        accy = []
        accz = []

        for idx in label_idxs:
            if accx is not None:
                accx = np.hstack((accx,total_acc_x.loc[idx][64:]))
                accy = np.hstack((accy,total_acc_y.loc[idx][64:]))
                accz = np.hstack((accz,total_acc_z.loc[idx][64:]))

            else:
                accx = total_acc_x.loc[idx]
                accy = total_acc_y.loc[idx]
                accz = total_acc_z.loc[idx]

        # saving the data into csv file
        data = pd.DataFrame({'accx':accx,'accy':accy,'accz':accz})
        save_path = os.path.join("Combined","Train",ACTIVITIES[label],f"Subject_{subject}.csv")
        data.to_csv(save_path,index=False)

print("Done Combining the training data")


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                        # Combining Test Data               
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Load all the accelerometer data
total_acc_x = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_x_test.txt"),delim_whitespace=True,header=None)
total_acc_y = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_y_test.txt"),delim_whitespace=True,header=None)
total_acc_z = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_z_test.txt"),delim_whitespace=True,header=None)

# Read the subject IDs
subject_test = pd.read_csv(os.path.join(test_path,"subject_test.txt"),delim_whitespace=True,header=None)

# Read the labels
y = pd.read_csv(os.path.join(test_path,"y_test.txt"),delim_whitespace=True,header=None)

# Toggle through all the subjects.
for subject in np.unique(subject_test.values):
    
        sub_idxs = np.where( subject_test.iloc[:,0] == subject )[0]
        labels = y.loc[sub_idxs]

        # Toggle through all the labels.
        for label in np.unique(labels.values):
    
            if not os.path.exists(os.path.join("Combined","Test",ACTIVITIES[label])):
                os.makedirs(os.path.join("Combined","Test",ACTIVITIES[label]))
    
            label_idxs = labels[labels.iloc[:,0] == label].index
    
            accx = []
            accy = []
            accz = []
            for idx in label_idxs:
                if accx is not None:
                    accx = np.hstack((accx,total_acc_x.loc[idx][64:]))
                    accy = np.hstack((accy,total_acc_y.loc[idx][64:]))
                    accz = np.hstack((accz,total_acc_z.loc[idx][64:]))
    
                else:
                    accx = total_acc_x.loc[idx]
                    accy = total_acc_y.loc[idx]
                    accz = total_acc_z.loc[idx]
    
            # saving the data into csv file
            data = pd.DataFrame({'accx':accx,'accy':accy,'accz':accz})
            save_path = os.path.join("Combined","Test",ACTIVITIES[label],f"Subject_{subject}.csv")
            data.to_csv(save_path,index=False)

print("Done Combining the testing data")
print("Done Combining the data")

#-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
```

```python
import pandas as pd

ACTIVITIES = ['walking', 'laying', 'sitting', 'walking_upstairs', 'walking_downstairs', 'standing']

# Initialize Groq model
from langchain_groq.chat_models import ChatGroq

Groq_Token = "API"  
model_name = "llama3-8b"  
groq_models = {
    "llama3-70b": "llama3-70b-8192", 
    "mixtral": "mixtral-8x7b-32768", 
    "gemma-7b": "gemma-7b-it",
    "llama3.1-70b": "llama-3.1-70b-versatile",
    "llama3-8b": "llama3-8b-8192",
    "llama3.1-8b": "llama-3.1-8b-instant",
    "gemma-9b": "gemma2-9b-it"
}
llm = ChatGroq(model=groq_models[model_name], api_key=Groq_Token, temperature=0)

# Example data for Few-Shot Learning
few_shot_examples = [
    {"mean_x": 0.997061, "mean_y": -0.263883, "mean_z": -0.073389, "std_x": 0.234325, "std_y": 0.193841, "std_z": 0.140181, "activity": "walking"},
    {"mean_x": 0.962942, "mean_y": -0.347216, "mean_z": -0.075989, "std_x": 0.210410, "std_y": 0.173446, "std_z": 0.174092, "activity": "laying"},
    {"mean_x": 0.920397, "mean_y": -0.230512, "mean_z": 0.322026, "std_x": 0.067967, "std_y": 0.111291, "std_z": 0.072892, "activity": "sTANDING"}
]

# Function to classify the activities using Few-Shot Learning with the Groq model
def classify_few_shot(row):
    few_shot_prompts = "\n".join([
        f"Activity: {example['activity']}, Mean x={example['mean_x']:.2f}, Mean y={example['mean_y']:.2f}, Mean z={example['mean_z']:.2f}, Std x={example['std_x']:.2f}, Std y={example['std_y']:.2f}, Std z={example['std_z']:.2f}."
        for example in few_shot_examples
    ])
    
    query = f"""
    * You are a classification model for human activities based on accelerometer data.
    * Here are some examples:
    {few_shot_prompts}
    * Now classify the following data into one of the six activities: 'walking', 'laying', 'sitting', 'walking_upstairs', 'walking_downstairs', 'standing'.
    * Data: Mean x={row['mean_x']:.2f}, Mean y={row['mean_y']:.2f}, Mean z={row['mean_z']:.2f}, Std x={row['std_x']:.2f}, Std y={row['std_y']:.2f}, Std z={row['std_z']:.2f}.
    * Provide only the activity name as your response, without any explanation.
    """

    # Get the model's prediction
    answer = llm.invoke(query)
    predicted_activity = answer.content.strip()
    
    return predicted_activity

# Apply the classification function to the DataFrame
test_stats_df['predicted_activity_few_shot'] = test_stats_df.apply(classify_few_shot, axis=1)

# Print predictions to verify
print(test_stats_df[[true_column, 'predicted_activity_few_shot']])
```

3. What are the limitations of Zero-Shot Learning and Few-Shot Learning in the context of classifying human activities based on featurized accelerometer data?

Zero-Shot Learning (ZSL) Limitations:

--> It depends on each task having an understandable and practical explanation. In the case that these descriptions are insufficient, the model may incorrectly categorize the actions.
--> It doesn't utilize any training examples for the tasks. This means that the model may have trouble with tasks for which it lacks sufficient knowledge or hasn't faced them before.
--> Sometimes, the actions that are described in Zero shot Learning vary from what is really done. This difference may result in incorrect categorization.
--> The point to which the descriptions contain all possible activities determines the model's capacity to accurately categorize new activities. The model may not work properly if the descriptions fail to fully explain the actions.

Few-Shot Learning (FSL) Limitations:

--> To acquire new actions, it just needs to see a few examples. The model might not pick up new information effectively if these examples are of low quality or are not representative.
--> If there are few examples, the model may become overly biased toward those examples and be less able to generalize to new cases of the same behavior.
--> There may be more examples for certain activities than for others. The model may perform weakly on tasks with fewer examples as a result of this imbalance.
--> It can be challenging to modify the model when there are few cases, and it could take some careful adjusting to get satisfactory outcomes.

4.What does the model classify when given input from an entirely new activity that it hasn't seen before?

```python
import pandas as pd
from scipy.spatial import distance

# Known activities with example mean feature vectors (replace with your actual data)
reference_data = pd.DataFrame({
    'activity': ['walking', 'laying', 'sitting', 'walking_upstairs', 'walking_downstairs', 'standing'],
    'mean_x': [0.4, 0.5, 0.3, 0.6, 0.5, 0.4],
    'mean_y': [0.3, 0.2, 0.4, 0.5, 0.3, 0.2],
    'mean_z': [0.5, 0.4, 0.3, 0.6, 0.5, 0.4],
    'std_x': [0.1, 0.15, 0.2, 0.1, 0.15, 0.1],
    'std_y': [0.2, 0.15, 0.2, 0.25, 0.2, 0.2],
    'std_z': [0.1, 0.2, 0.15, 0.1, 0.2, 0.15]
})

# New activity data to classify
new_activity_data = pd.DataFrame({
    'mean_x': [0.5],
    'mean_y': [0.3],
    'mean_z': [0.4],
    'std_x': [0.1],
    'std_y': [0.15],
    'std_z': [0.2]
})

def classify_activity(row):
    # Calculate the Euclidean distance between the new data and each reference activity
    distances = []
    for _, ref_row in reference_data.iterrows():
        dist = distance.euclidean(
            [row['mean_x'], row['mean_y'], row['mean_z'], row['std_x'], row['std_y'], row['std_z']],
            [ref_row['mean_x'], ref_row['mean_y'], ref_row['mean_z'], ref_row['std_x'], ref_row['std_y'], ref_row['std_z']]
        )
        distances.append(dist)
    
    # Find the index of the minimum distance
    min_index = distances.index(min(distances))
    return reference_data.iloc[min_index]['activity']

# Apply the classification to the new activity data
new_activity_data['predicted_activity'] = new_activity_data.apply(classify_activity, axis=1)

# Print the results
print(new_activity_data[['mean_x', 'mean_y', 'mean_z', 'std_x', 'std_y', 'std_z', 'predicted_activity']])
```

When the model is provided with input from an entirely new activity that it has not seen before, it will classify the activity based on the nearest known activity from the reference dataset. The model calculates the Euclidean distance between the new activitys features and the features of each known activity. It then assigns the new activity to the category of the closest known activity in terms of distance. This approach does not account for the actual distinctiveness of the new activity but relies on the similarity to the most similar known activity in the reference set. Consequently, if the new activity is significantly different from all known activities, it may still be classified into a category that is closest in feature space, potentially leading to a misclassification if the new activity has unique characteristics not captured by the reference data.
